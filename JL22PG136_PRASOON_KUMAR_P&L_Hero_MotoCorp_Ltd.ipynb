## Statistics Modeling

```python
import statsmodels.api as sm
x = sm.add_constant(data[['Excise Duty', 'Net Sales', 'Other Income', 'Stock Adjustments', 'Total Income', 'Raw Materials', 'Employee Cost', 'Other Manufacturing Expenses', 'Selling and Administration Expenses', 'Miscellaneous Expenses', 'Total Expenditure', 'Operating Profit', 'Gross Profit', 'Depreciation', 'Profit Before Tax', 'Tax', 'Deferred Tax', 'Reported Net Profit', 'Extraordinary Items', 'Adjusted Net Profit', 'Adjst. below Net Profit', 'P & L Balance brought forward', 'Appropriations', 'P & L Balance carried down', 'Dividend', 'Equity Dividend %', 'Dividend Per Share(Rs)', 'Earnings Per Share-Unit Curr', 'Earnings Per Share(Adj)-Unit Curr', 'Book Value-Unit Curr', 'Book Value(Adj)-Unit Curr', 'Dividend Per Share Adj.(Rs)']])
x.head()
```

```python
y = data['Year']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=100)
x_train
```

```python
from IPython.display import display
data_lm = sm.OLS(y_train, x_train).fit()
display( data_lm.params )
```

### R2

```python
data_lm.summary()
```

**The output is the result of a linear regression model with 'Year' as the dependent variable and several independent variables as the independent variables. However, the output reveals model flaws, and it is challenging to derive meaningful conclusions from the data. Here are a few of the problems:**

**	The R-squared value is 1, indicating that the model fully explains the variance in the dependent variable 'Year.' This value is excessive and suggests that the model may be overfitting.**


**	The adjusted R-squared value is 'nan,' indicating that the model may contain too many variables relative to the number of observations, resulting in an unstable or invalid model.**

**	The F-statistic and its associated probability are both 'nan,' indicating that the overall significance of the model cannot be ascertained.**

**	The coefficients of all independent variables are either zero or 'nan,' and their corresponding t-values and p-values are also 'nan.' This indicates that the independent variables have no effect on the dependent variable, and their incorporation in the model may be superfluous.**

**	The coefficients' standard errors are 'inf,' indicating that there may be numerical errors in the model, potentially as a result of data scaling or transformation.**

**	The results of the Omnibus and Jarque-Bera tests for normality indicate that the residuals of the model may not be normally distributed, casting further doubt on the validity of the model.**

**	The Durbin-Watson test for autocorrelation indicates that the model's assumption of independent errors may be flawed.**

```python
data_resid = np.random.normal(size=100)

probplot = sm.ProbPlot(data_resid)
fig = probplot.ppplot(line='45')

plt.title("Normal P-P Plot of Hero MotoCorp Regression Standardized")

plt.show()
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
resid = data_resid[data_lm.fittedvalues.index]
```

```python
# Plot the scatter plot
plt.scatter(get_standardized_values(data_lm.fittedvalues), get_standardized_values(resid))
plt.title("Residual Plot: Hero P&L Prediction")
plt.xlabel("Standardized predicted values")
plt.ylabel("Standardized residuals")
plt.show()
```
